```python
from matplotlib import pyplot as plt
import numpy as np
import sympy as sym
import math 
import random
```

Here, we define the function we need to integrate, as well as the integration range.

```python
#define our function
x=sym.Symbol('x');
f=x**2

#integration range
a=0;
b=1;
```

We can now implement Monte Carlo's algorithm. We need to take as arguments the dimensions of the rectangle in which we operate, as well as the number of points we want to define.
You can first implement a function that generates random points, then stores them in different lists depending on whether they are in the area under the curve. We will also plot these points, in order to visualise the integration.

```python
def montecarlo(f1,ax1,ax2,ay1,ay2,n):
    L0=[]; L1=[]; L_1=[]
    #First, we generate n random points
    for i in range(n):
        ax=random.uniform(ax1,ax2);
        ay=random.uniform(ay1,ay2);
        #We place each random point in L1 or L2, depending on whether it belongs to the integration area
        if (ay<f1.subs(x,ax) and ay>0):
            L1.append((ax,ay));
        else:
            if (ay>f1.subs(x,ax) and ay<0):
                L_1.append((ax,ay));
            else: L0.append((ax,ay));
                
    I=((float(len(L1)-len(L_1))/n)*float((ax2-ax1)*(ay2-ay1)));
    Lx1, Ly1 = zip(*L1);
    
    #We can now plot the lists
    plt.plot(Lx1, Ly1,marker='o', linestyle='None', markersize = 5.0);
    if L_1!=[]:
        Lx_1, Ly_1 = zip(*L_1);
        plt.plot(Lx_1, Ly_1,marker='o', linestyle='None', markersize = 5.0);
    Lx0, Ly0 = zip(*L0);
    plt.plot(Lx0, Ly0,marker='o', linestyle='None', markersize = 5.0);
    plt.show;
    print(I);
```

We can now run the algorithm and visualise the results.

```python
montecarlo(f,0,3,-1,9,5000);
```
